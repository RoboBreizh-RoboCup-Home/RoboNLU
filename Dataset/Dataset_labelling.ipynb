### Phrase Driver

```python
def divide_sentence_in_phrases(sentence):
    a = ''
    sentence = sentence.split()
    
    for k, i in enumerate(sentence):
        if i in all_verb_lst and k != 0:
            for j in all_verb_lst:
                if i == j:
                    a += ',' + i
        else:
            a += ' '+i

    return a.replace(' ,',',').replace(', and',',').replace('and ', ',').replace(',,', ',').replace(' and,',',').replace('.','').split(',')
```

```python
def phrase_driver(process_df):
    divided_lst = []
    count = 0
    com_lst = []
    for idx in range(len(process_df)):
        #print(idx+1)
        com = list(process_df['Commands'])[idx]
        com_lst.append(com)

        sub_task_lst = divide_sentence_in_phrases(com)
        sub_task_lst = [task.strip() for task in sub_task_lst]
        
#         if idx == 100:
#             return

        offset_lst = []
        #print(com)
        for task in sub_task_lst:
            #print(task.split())
            offset = com.split().index(task.split()[0])
            offset_lst.append(offset)

        sub_task_dict = {}
        for i in range(len(sub_task_lst)):
            if sub_task_lst[i].split()[0] in all_verb_lst:
                k = sub_task_lst[i]
                sub_task_dict.update({k:{'B_idx':offset_lst[i]}})
                
        if len(sub_task_dict) == 0:
            count += 1
            divided_lst.append([com,{None: None}])
        else:
            divided_lst.append([com,sub_task_dict.copy()])
                               
            
    return divided_lst


divided_lst = phrase_driver(process_df)
```

```python
pronoun_lst = ['me','you','he','she','him','her','it','them','us','this','that','these','those']

slot_dict = {'obj':objects_lst, 'dest': destination_lst, 'per':name_lst, 'pro':pronoun_lst}
```

```python
def increasing_len_entity(slot_dict):
    lst = []
    for i in range(len(slot_dict)):
        lst.extend(list(slot_dict.values())[i])

    lst.sort(key = len)
    
    for i,ele in enumerate(lst):
        for categ,word_lst in slot_dict.items():
            if ele in word_lst:
                lst[i] = (ele,categ)
    return lst

incre_entity_lst = increasing_len_entity(slot_dict)
incre_entity_lst
```

```python
def slot_labelling(lst,pronoun_lst,slot_dict,incre_entity_lst):
    
    #slotted_dict = intented_lst.copy()
    intented_lst = lst.copy()
    slotted_lst = []
    for item in intented_lst:
        com,sub_task_dict = item[0],item[1]
        #print(com)

        for task,info_dict in sub_task_dict.items():
            #print('task: ', task)
            slot_info = {}
            if task == None:
                continue
            
            for entity,categ in incre_entity_lst:
                begin1 = task.find(' '+entity+' ')
                begin2 = task.find(entity)
                
                slot_key = categ
                
                if begin1 != -1:
                    w_idx = task.split().index(entity.split()[0])
                    #slot_info.update({slot_key:[begin2,entity]})
                    slot_info.update({slot_key:[w_idx,entity]})

                    #print('!!store 1')
                elif begin2 != -1 :
                    if task.find(entity+' ') != -1 and begin2 == 0:
                        w_idx = task.split().index(entity.split()[0])
                        slot_info.update({slot_key:[w_idx,entity]})
                        #slot_info.update({slot_key:[begin2,entity]})
                        #print('!!store 2' )
                    elif task.find(' '+entity) != -1 and begin2 + len(entity) == len(task):
                        w_idx = task.split().index(entity.split()[0])
#                         print(entity)
#                         print(w_idx)
#                         print(entity.split()[0])
                        slot_info.update({slot_key:[w_idx,entity]})


            info_dict.update({'slot': slot_info})
        
        slotted_lst.append([com,sub_task_dict])
    return slotted_lst
```

```python
intent_label_lst = ['go',
             'find',
             'follow',
             'take',
             'put',
             'say']



print('intent_label_lst: ',len(intent_label_lst))


slot_categ_lst = ['obj','per','dest','pro']

def intent_slot_label(intent_label_lst,slot_categ_lst):
    inten_slot_label_dict = {}
    for inten in intent_label_lst:
        dic = {}
        inten_slot_label_dict.update({inten:dic})
        for s_categ in slot_categ_lst:
            dic.update({s_categ:['B-'+inten+'.'+s_categ,'I-'+inten+'.'+s_categ]})
            
    return inten_slot_label_dict
```

```python
def create_labels(process_df, slotted_lst,inten_slot_label_dict):
    labeled_df = process_df.copy()
    labeled_df['Label'] = [None]*len(labeled_df)
    labeled_df['Label'] = labeled_df['Label'].astype('object')
    #lst_all = []
    for i,item in enumerate(slotted_lst):
        com,sub_task_dict = item[0],item[1]
        #temp = []
        lst = []
        #print(com,'\n')
#         if i == 941:
#             print(i)
#             print(com)
#             print(sub_task_dict)
        for sub_task, task_info_dict in sub_task_dict.items():
            if sub_task == None:
                continue

            for k,v in task_info_dict.items():
                if k == 'B_idx':
                    offset = v
                    
                elif k == 'intent':
                    inten = v
                    lst.append((v,offset))
                    
                elif k == 'slot':
                    for categ, entity_info_lst in v.items():
                        entity_B_idx = entity_info_lst[0]
                        entity = entity_info_lst[1]
                        word_len = len(entity.split())
                        for t in range(word_len):
                            idx = offset + entity_B_idx + t
                            if t == 0:
                                label = inten_slot_label_dict[inten][categ][0]
                                lst.append((label,idx))
                            else:
                                label = inten_slot_label_dict[inten][categ][1]
                                lst.append((label,idx))
                
        #print(lst)
        labeled_df['Label'] = labeled_df['Label'].astype('object')
#         print(i)
#         print(lst)
        labeled_df.at[i, 'Label'] = lst
        #display(labeled_df)
        #print('-----------------------------------')
        #lst_all.append(lst)
    return labeled_df
```

```python
def Generate_Dataset(labeled_df,process_df):
    dataset_df = labeled_df.copy()
    dataset_df['seq_out'] = [None]*len(process_df)
    dataset_df['seq_out'] = dataset_df['seq_out'].astype('object')
    
    for i in range(len(labeled_df)):
        label = dataset_df['Label'].iloc[i]
        com = dataset_df['Commands'].iloc[i]
        seq = [None]*len(com.split())
        
        for lab,idx in label:
            seq[idx] = lab
        
        dataset_df.at[i, 'seq_out'] = ['O' if element == None else element for element in seq ]
    return dataset_df
    
dataset_df = Generate_Dataset(labeled_df,process_df)
dataset_df
```

## Few-shot Learning --  for labelling

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli") ## Few-shot Bert
```

```python
def majority_vote(verb_lst,verb_type_dict):
    votes = {key:0 for key in verb_type_dict.keys()}

    for verb in verb_lst:
        for k,v in verb_type_dict.items():
            if verb in v:
                votes[k] += 1
    
    return  max(votes, key=votes.get)
```
